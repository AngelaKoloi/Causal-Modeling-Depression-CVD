```python
def calculate_descriptives(df, round=2):
    descriptives = {}
    for column in df.columns:
        col_data = df[column]
        N = col_data.count()
        first_quartile = col_data.quantile(0.25)
        third_quartile = col_data.quantile(0.75)
        Min = col_data.min()
        Max = col_data.max()
        Mean = col_data.mean()
        Median = col_data.median()
        Mode = col_data.mode().iloc[0] if not col_data.mode().empty else np.nan
        SD = col_data.std()

        descriptives[column] = {
            "N": N,
            "1st Qu.": first_quartile,
            "3rd Qu.": third_quartile,
            "Min.": Min,
            "Max.": Max,
            "Mean": Mean,
            "Median": Median,
            "Mode": Mode,
            "SD": SD
        }
    df_descriptives = pd.DataFrame(descriptives).round(round)
    return df_descriptives
```

```python
def get_descriptives(df_input):
    df = df_input.copy()
    df = df.drop(df.index[0])
    df = calculate_descriptives(df)
    
    sort_list = list(new_column_names.values())
    df = df[sort_list]
    df = df.T
    
    df_filtered = df[df.index.str.endswith('_D')]
    for idx in df_filtered.index:
        df.loc[idx, "Mean"] = "-"

    df["N"] = df["N"].astype(int)
    
    return df
```

```python
df_plot = get_descriptives(df)
df_plot[~df_plot.index.str.endswith('_CO')]
```

```python
def plot_descriptives(df, image_name="../images/mytable.png"):
    ax = plt.subplot(111, frame_on=False) 
    ax.xaxis.set_visible(False)
    ax.yaxis.set_visible(False)
    
    
    table = ax.table(cellText=df.values,
                     colLabels=df.columns,
                     rowLabels=df.index,
                     loc='center',
                     cellLoc='center')
    
    plt.savefig(image_name, bbox_inches='tight', pad_inches=0)
    plt.close()
```

```python
split_index = (len(df_plot) // 2)

df1 = df_plot.iloc[:split_index]
df2 = df_plot.iloc[split_index:]
df_list = [df1, df2]

for count, df in enumerate(df_list):
    plot_descriptives(df, f"../images/descriptive_{count+1}.png")
```
